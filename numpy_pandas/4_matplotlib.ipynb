```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 기본 lineplot

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
# 그래프를 그리는 코드 작성
fig,ax=plt.subplots()
plt.plot(x,y)
ax.plot(x,y)
ax.set_title("First Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

## 심화 lineplot

```python
x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(
    x, x, label='y=x',
    marker='o',
    color='blue',
    linestyle=':'
)
ax.plot(
    x, x**2, label='y=x^2',
    marker='^',
    color='red',
    linestyle='--'
)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(
    loc='best',
    shadow=True,
    fancybox=True,
    facecolor= "grey",
    borderpad=2

)
```

## Scatter

```python
fig, ax = plt.subplots()
x = np.random.randn(50)
y = np.random.randn(50)
colors = np.random.randint(0, 100, 50)
sizes = 500 * np.pi * np.random.rand(50) ** 2

ax.scatter(
    x, y, c = colors, s = sizes
)
```

## Bar & Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# fname='./NanumBarunGothic.ttf'
# font = fm.FontProperties(fname = fname).get_name()
# plt.rcParams["font.family"] = font

plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False)

x = np.array(["축구", "야구", "농구", "배드민턴", "탁구"])
y = np.array([18, 7, 12, 10, 8])

z = np.random.randn(1000)


fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Bar 그래프
axes[0].bar(x, y)
# 히스토그램
axes[1].hist(z, bins = 50)
```
